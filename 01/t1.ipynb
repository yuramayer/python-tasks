Найдём количество элементов в ней

Мы будем проходиться по последовательности, 
    шагами по длина all_elems, и двигаться на +1 индекс вперёд 
        до конца строки

Каждый раз мы будем проверять:

- длина множества по такая же как у all_elems?

> если длина множества такая же, значит все элементы разные
>
> значит нет дублей, и при этом есть все элементы
>
> и длина all_elems = константа

```python

# итерируемся по СЕМЁРКАМ = длине set_len
# каждый раз добавляем единичку, тк идём "направо" по строке
# в конце пришлось прибавить единичку, 
#   хз почему - без неё почему-то не хватало последней строки

# если нашли схожий по длине множеств - break

for i in range(string_len - set_len + 1):
    part_set = set(string[i : i + set_len])
    if len(part_set) == set_len:
        print('found:', string[i : i + set_len])
        break 
```

```python
def is_unique_substring(string):
    string_len = len(string)

    # здесь собрали уникальные символы,
    # перебрали все len(string) симолов
    # Время: операция заняла O(n),
    # Память: заняли O(k), где k - число символов k <= n
    all_elems = set(string)
    set_len = len(all_elems)

    # повторений в цикле примерно n - k + 1 => O(n) раз
    for i in range(string_len - set_len + 1):

        # на каждой итерации мы делаем срез длины k => O(k)
        # так как копируем подстроку длиной k
        # и проводим для этой строки set() => снова O(k)

        # итого O(k + k) => O(k)
        part_set = set(string[i : i + set_len])
        if len(part_set) == set_len:
            return True
        
        # всего по циклу сложность уровня 
        # O((n - k + 1)) * k) 
        # n - k + 1 <= n, (n - k + 1) * k = n * k
        # значит сложность уровня O(n * k)
    return False

# в худшем случае k приблизтельно равно n,
# поэтому сложность по худшему сценарию становится O(n^2)

# с памятью лучше, она требует O(n)
```
